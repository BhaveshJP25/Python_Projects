```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Image.jpeg")
plt.imshow(img)
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray)
```

```python
inveredImg = 255 - imgGray
plt.imshow(inveredImg)
```

```python
invertedGaussianSmoothing = cv2.GaussianBlur(inveredImg, (21, 21), sigmaX=0, sigmaY=0)
plt.imshow(invertedGaussianSmoothing)
```

```python
greyScaleGaussianSmoothing = 255 - invertedGaussianSmoothing
plt.imshow(greyScaleGaussianSmoothing)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

sketch = cv2.divide(imgGray, greyScaleGaussianSmoothing, scale=256)
ax1.imshow(img)
ax2.imshow(sketch)
```

```python
cv2.imshow("Sketch", sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
