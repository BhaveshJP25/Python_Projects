```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Image.jpeg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(imgGray, cv2.COLOR_BGR2RGB))
```

```python
invertedImg = 255 - imgGray
plt.imshow(cv2.cvtColor(invertedImg, cv2.COLOR_BGR2RGB))
```

```python
invertedGaussianSmoothing = cv2.GaussianBlur(invertedImg, (21, 21), sigmaX=0, sigmaY=0)
plt.imshow(cv2.cvtColor(invertedGaussianSmoothing, cv2.COLOR_BGR2RGB))
```

```python
greyScaleGaussianSmoothing = 255 - invertedGaussianSmoothing
plt.imshow(cv2.cvtColor(greyScaleGaussianSmoothing, cv2.COLOR_BGR2RGB))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

sketch = cv2.divide(imgGray, greyScaleGaussianSmoothing, scale=256)
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
```

```python
cv2.imshow("Sketch", sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
