```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Image.jpeg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(imgGray, cv2.COLOR_BGR2RGB))
```

```python
gaussianSmoothing = cv2.GaussianBlur(imgGray, (21, 21), sigmaX=0, sigmaY=0)
plt.imshow(cv2.cvtColor(gaussianSmoothing, cv2.COLOR_BGR2RGB))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

sketch = cv2.divide(imgGray, gaussianSmoothing, scale=256)
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
```

```python
cv2.imshow("Sketch", sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
